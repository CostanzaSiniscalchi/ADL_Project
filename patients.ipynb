```python
import os
import numpy as np
import re
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
base_dir = 'data_t1w_3_scans/'

# Initialize an empty list to store data
data = []
data2 = []
```

```python
# Iterate over each patient's directory
for patient_dir in os.listdir(base_dir):
    patient_path = os.path.join(base_dir, patient_dir)
    if os.path.isdir(patient_path):
        
        # Iterate over each scan directory
        for scan_dir in os.listdir(patient_path):
            scan_path = os.path.join(patient_path, scan_dir)
            initial_age = None
            print(scan_path)
            if scan_path.endswith('.json'):
                
                with open(scan_path, 'r') as f:
                    json_content = f.read()

                # Fix the single quotes by replacing them with double quotes
                json_content = json_content.replace("'", '"').replace(": None", ': null')
                
                
                # Now load the fixed JSON content
                try:
                    scan_data = json.loads(json_content)
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON from {scan_path}: {e}")
                    continue
                
                # Extract relevant fields from the JSON
                m_cand_id = scan_data.get('CandID', None)
                m_gender = scan_data.get('Gender', None)
                m_language = scan_data.get('Language', None)
                m_fname = scan_path

            if os.path.isdir(scan_path):
                # print(os.listdir(scan_path))
                # Iterate over each file in the scan directory
                for file_name in os.listdir(scan_path):
                    file_path = os.path.join(scan_path, file_name)

                    # Check if the file name matches the pattern for JSON files
                    if file_name.endswith('.json'):
                        with open(file_path, 'r') as f:
                            json_content = f.read()

                        # Fix the single quotes by replacing them with double quotes
                        json_content = json_content.replace("'", '"')
                        
                        # Now load the fixed JSON content
                        try:
                            scan_data = json.loads(json_content)
                        except json.JSONDecodeError as e:
                            print(f"Error decoding JSON from {file_path}: {e}")
                            continue
                        
                        # Extract relevant fields from the JSON
                        meta = scan_data.get('Meta', {})
                        cand_id = meta.get('CandID', None)
                        visit = meta.get('Visit', None)
                        file_name_json = meta.get('File', None)
                        qc = scan_data.get('QC', None)
                        selected = scan_data.get('Selected', None)
                        caveats = scan_data.get('Caveats', [])
                        
                        if file_name == 'session.json' and scan_dir == 'PREBL00':
                            age = scan_data.get('Age_at_MRI', None)
                            print(age)
                            
                        # Process the caveats if needed (just counting them here)
                        caveat_count = len(caveats)
                        
                        # Append the data to the list
                        data.append({
                            'PatientID': patient_dir,
                            'ScanID': scan_dir,
                            'CandID': cand_id,
                            'Visit': visit,
                            'File': file_name_json,
                            'QC': qc,
                            'Selected': selected,
                            'CaveatCount': caveat_count,
                            'FileName': file_name
                        })
        # Append the data to the list
        data2.append({
            'CandID': m_cand_id,
            'File': m_fname,
            'Gender': m_gender,
            'Language': m_language,
            'Initial Age in Months': age,
            'Initial Age in Years': age//12
        })        

# Convert the list of dictionaries into a pandas DataFrame
df = pd.DataFrame(data)
df2 = pd.DataFrame(data2)

df.to_csv('preventad_patients.csv')
df2.to_csv('preventad_candidates.csv')
df
```

```python
data = []
# Iterate over each patient's directory
for patient_dir in os.listdir(base_dir):
    patient_path = os.path.join(base_dir, patient_dir)
    if os.path.isdir(patient_path):
        
        # Iterate over each scan directory
        for scan_dir in os.listdir(patient_path):
            scan_path = os.path.join(patient_path, scan_dir)
            if os.path.isdir(scan_path):
                # print(os.listdir(scan_path))
                # Iterate over each file in the scan directory
                for file_name in os.listdir(scan_path):
                    file_path = os.path.join(scan_path, file_name)
                    numfiles = len([file for file in os.listdir(scan_path) if not file.endswith('.json')])
                    # Check if the file name matches the pattern for JSON files
                    if not file_name.endswith('.json'):
                        
                        # Append the data to the list
                        data.append({
                            'PatientID': patient_dir,
                            'ScanID': scan_dir,
                            'FileName': file_name,
                            'NumFilesInScan': numfiles
                        })
                    

# Convert the list of dictionaries into a pandas DataFrame
df = pd.DataFrame(data)

df.to_csv('preventad_filestats.csv')
df
```
