```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv = 'data-2025-03-21T14_24_34.904Z.csv'

df = pd.read_csv(csv)
print(df.columns)
df
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Age At MRI In Months'].value_counts().index, df['Age At MRI In Months'].value_counts().values)
plt.xlabel('Age at MRI (months)')
plt.ylabel('Frequency')
plt.title('Distribution of Age at MRI (in months)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Handedness Score'].value_counts().index, df['Handedness Score'].value_counts().values)
plt.xlabel('Handedness Score')
plt.ylabel('Frequency')
plt.title('Distribution of Handedness Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Gender'].value_counts().index, df['Gender'].value_counts().values)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Distribution of Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Handedness Interpretation'].value_counts().index, df['Handedness Interpretation'].value_counts().values)
plt.xlabel('Handedness Interpretation')
plt.ylabel('Frequency')
plt.title('Distribution of Handedness Interpretation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
dates = pd.to_datetime(df['First Insertion'])

plt.figure(figsize=(10, 6))
plt.bar(dates.value_counts().index, dates.value_counts().values)
plt.xlabel('First Insertion Date')
plt.ylabel('Frequency')
plt.title('Distribution of Dates')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
